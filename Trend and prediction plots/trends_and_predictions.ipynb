```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the CSV file
df = pd.read_csv('africa_poverty_cleaned.csv')

# Filter for a specific country (e.g., Nigeria)
country = "Kenya"
country_df = df[df["Country Name"] == country].sort_values("Year")

# Prepare data for regression
X = country_df["Year"].values.reshape(-1, 1)
y = country_df["Poverty Rate"].values

# Fit linear regression
model = LinearRegression()
model.fit(X, y)

# Predict up to 2030
future_years = np.arange(country_df["Year"].min(), 2031).reshape(-1, 1)
predictions = model.predict(future_years)

# Plot actual and predicted
plt.figure(figsize=(10, 6))
plt.plot(country_df["Year"], country_df["Poverty Rate"], marker="o", label="Actual Poverty Rate")
plt.plot(future_years, predictions, linestyle="--", color="red", label="Predicted Poverty Rate (to 2030)")
plt.title("Kenya Poverty Rate Trend and Prediction")
plt.xlabel("Year")
plt.ylabel("Poverty Rate (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("screenshots.png")
plt.show()
```
